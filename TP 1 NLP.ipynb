```python
corpus = pd.concat([jeu_apprentissage, jeu_test])
```

```python
mots_apprentissage = ' '.join(corpus['Contenu']).split()
lexique_apprentissage = set(mots_apprentissage)
```

```python
lexique_apprentissage
```

```python
numerotation = {mot: num + 1 for num, mot in enumerate(lexique_apprentissage)}
```

```python
numerotation
```

```python
jeu_apprentissage['Sentiment'].unique()
```

```python
correspondance_labels = {
    'objective': 0,
    'positive': 1,
    'negative': 2,
    'mixed': 3
}
```

```python
def creer_ligne(tweet, numerotation, correspondance_labels):
    mots = tweet.split()  
    lignes = {}
    for mot in mots:
        if mot in numerotation:
            index = numerotation[mot]
            if index not in lignes:
                lignes[index] = 0
            lignes[index] += 1
    sorted_lines = sorted([f"{index}:{count}" for index, count in lignes.items()], key=lambda x: int(x.split(":")[0]))
    return " ".join(sorted_lines)
```

```python
with open("train.svm", "w") as f:
    for index, row in jeu_apprentissage.iterrows():
        label = correspondance_labels[row['Sentiment']]
        tweet = row['Contenu']
        ligne = creer_ligne(tweet, numerotation, correspondance_labels)
        f.write(f"{label} {ligne}\n")
```

```python
with open("test.svm", "w") as f:
    for index, row in jeu_test.iterrows():
        label = correspondance_labels[row['Sentiment']]
        tweet = row['Contenu']
        ligne = creer_ligne(tweet, numerotation, correspondance_labels)
        f.write(f"{label} {ligne}\n")
```
